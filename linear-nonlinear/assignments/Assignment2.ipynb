### 3
#### Given link function is logit, compute and plot the mean response against the defined linear predictor

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the linear predictor range
linear_predictor = np.arange(-5, 5.01, 0.01)

# Calculate the mean response using the logistic function
# mean_response = 1 / (1 + np.exp(-linear_predictor))
mean_response = np.exp(linear_predictor)/(1 + np.exp(linear_predictor))

# Plot the mean response against the linear predictor
plt.figure(figsize=(10, 6))
plt.plot(linear_predictor, mean_response, label=r'$\mu_i = \frac{e^{x_i^T\beta}}{1+e^{x_i^T\beta}}$')
plt.title('Mean Response vs. Linear Predictor')
plt.xlabel(r'Linear Predictor, $x_i^T\beta$')
plt.ylabel(r'Mean Response, $\mu_i$')
plt.legend()
plt.grid(True)
plt.show()
```

### 6
#### Show likelihood varying over different values of lambda and show MLE of lambda for which the likelihood is the highest.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(r'/Users/xnxk040/Library/CloudStorage/OneDrive-W.WGrainger,inc/Desktop/linear-nonlinear data/poisson_mle.csv')
column_name = 'data'

# Define the lambda sequence
lambdas = np.arange(1, 21, 0.01)

# Calculate likelihood for each lambda
likelihoods = []

for lamb in lambdas:
    # Calculate the Poisson PMF and then the product of those for the likelihood
    likelihood = np.prod([(lamb**x) * np.exp(-lamb) / np.math.factorial(x) for x in data[column_name]])
    likelihoods.append(likelihood)

# Find the lambda that maximizes the likelihood
max_likelihood_index = np.argmax(likelihoods)
mle_lambda = lambdas[max_likelihood_index]

# Plot the likelihoods
plt.figure(figsize=(10, 6))
plt.plot(lambdas, likelihoods, label='Likelihood')
plt.axvline(x=mle_lambda, color='r', linestyle='--', label=f'MLE of Lambda: {mle_lambda:.2f}')
plt.title('Likelihood of Different Lambda Values in Poisson Distribution')
plt.xlabel('Lambda')
plt.ylabel('Likelihood')
plt.legend()
plt.grid(True)
plt.show()
```
